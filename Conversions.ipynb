```python
# Compound Interest
p = float(input("Enter Principle amount : "))
r = float(input("Enter Rate of interest % : "))
t = float(input("Enter Time in years : "))

A = (p * (1 + r/100)**t)
print("Amount is ",A)
CI = A-p

print("Compound Interest",round(CI,2))
```
