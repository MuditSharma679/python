### Deep and Shallow Copy
Shallow copy shares memory with the original array, while deep copy creates a completely independent copy.
Use shallow copy → save memory, read-only ops

Use deep copy → when you will modify data safely

```python
a = np.arange(1,21)
a
```

```python
# Deep Copy : - This slicing change does not happen in the actual array
# But happen in list
slice = a[:5]
slice = slice * 10
slice
```

```python
a
```

```python
a
b = a.copy()
```

```python
b[0] = 99
b
```

```python
a
```

### Advance Array Manipulation
Stacking Array, here we have to use vectors

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```
