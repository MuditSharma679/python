# Numpy Indexing and Slicing of Vectors

```python
import numpy as np
```

```python
arr = np.arange(11,21)
arr
```

```python
arr[8]
```

```python
arr[1:5] #Here we give index num start and end
```

```python
arr[:5]
```

```python
arr[5:]
```

```python
arr_slice = arr[3::2] #[Start:Stop:Step]
arr_slice
```

# Numpy Indexing and Slicing of Matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0,4]    #[row,column]
```

```python
slice = arr[0:2,1:3]
slice
```

```python
slice2 = arr[2:4,3:]
slice2
```

```python
slice3 = arr[3:,3:]
slice3
```

```python
arr[:,2]
```

```python
arr[::2, ::2]
```

```python
arr[arr%2==0]
```

# Boolean Indexing

```python
arr =np.arange(11,21)
arr
```

```python
bool_index = arr%2 == 0
bool_index
```

```python
arr2 = arr[bool_index]
arr2
```

```python
arr1 = arr[arr % 2 == 0]
arr1
```
