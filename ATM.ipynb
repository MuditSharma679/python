### Q.2 ATM Pin check - correct/incorrect, balance check, balance withdraw , minimum_balance = 1000.

```python
pin = int(input("Enter Your PIN:- "))
balance = 10000

if(pin == 1234):
    print("CORRECT ATM PIN")
    print("Your Balance: ",balance)
    withdraw_amount = int(input("Enter the Amount to Withdraw:- "))
    if(balance >= withdraw_amount):
        if(balance <= (withdraw_amount+1000)):
            print(withdraw_amount,"RS failed to withdraw because it is "+str(1000-(balance-withdraw_amount))+" more than minimum balance")
        else :
            balance = balance - withdraw_amount
            print(withdraw_amount,"RS Withdraw Successfully")
            print("Remaining Balance is:-",balance)
    
else:
    print("Incorrect PIN")
```

```python
pin = int(input("Enter Your PIN:- "))
balance = 20000
min_balance = 1000

if(pin == 1234):
    print("CORRECT ATM PIN")
    print("Your Balance:", balance)

    withdraw_amount = int(input("Enter the Amount to Withdraw:- "))

    if(withdraw_amount <= 0):
        print("Invalid withdrawal amount")

    elif(withdraw_amount > balance):
        print("Insufficient balance")

    elif(balance - withdraw_amount < min_balance):
        less_amount = min_balance - (balance - withdraw_amount)
        print(withdraw_amount,"RS failed to withdraw because it is",less_amount,"more than minimum balance")

    else:
        balance -= withdraw_amount
        print(withdraw_amount, "RS Withdraw Successfully")
        print("Remaining Balance is:-", balance)

else:
    print("Incorrect PIN")
```

Q.3) Change Password Program:-

```python
old_pin = int(input("Enter your old pin:- "))

if(old_pin == 1234):
    print("CORRECT OLD PIN")
    new_pin = int(input("Enter NEW PIN:- "))
    cfm_new_pin = int(input("Confirm NEW PIN:- "))
    if(new_pin == cfm_new_pin):
        print("Your PIN changed Successfully")
    else:
        print("Both PIN not matched")
        
else:
    print("INCORRECT OLD PIN")  
    
```
