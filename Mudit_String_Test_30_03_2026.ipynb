#### 1). isupper()

```python
s = input("Enter the string to check for uppercase: ")
has_upper = False

for i in s:
    ascii_val = ord(i)

    if(65 <= ascii_val <= 90):
        has_upper = True
    
    elif(97 <= ascii_val <= 122):
        print(False)
        break
else:
    # Runs when we don't find any lower case
    print(has_upper)
```

#### 2). islower()

```python
s = input("Enter the string to check for lowercase: ")
has_lower = False

for i in s:
    ascii_val = ord(i)

    if(97 <= ascii_val <= 122):
        has_lower = True
    
    elif(65 <= ascii_val <= 90):
        print(False)
        break
else:
    # Runs when we don't find any upper case
    print(has_lower)
```

#### 3). isdigit()

```python
s = input("Enter the string: ")

if s == '':
    print(False)
else:
    for i in s:
        ascii_val = ord(i)
        if not (48 <= ascii_val <= 57):
            print(False)
            break
    else:
        print(True)
```

#### 4). isspace()

```python
s = input("Enter the string: ")
L = 0

for i in s:
    L+=1
if L == 0:
    print(False)
else:
    for i in s:
        ascii_val = ord(i)
        if not(ascii_val == 32):
            print(False)
            break
    else:
        print(True)
```

#### 7). isalpha()

```python
s = input("Enter the string: ")
L = 0

for i in s:
    L+=1
if L == 0:
    print(False)
else:
    for i in s:
        ascii_val = ord(i)
        if not ((ascii_val >= 65 and ascii_val <= 90) or (ascii_val >= 97 and ascii_val <= 122)):
            print(False)
            break
    else:
        print(True)
```

#### 8). isalnum()

```python
s = input("Enter the string: ")
L = 0

for i in s:
    L+=1
if L == 0:
    print(False)
else:
    for i in s:
        ascii_val = ord(i)
        if not ((ascii_val >= 65 and ascii_val <= 90) or (ascii_val >= 97 and ascii_val <= 122) or (ascii_val >= 48 and ascii_val <= 57)):
            print(False)
            break
    else:
        print(True)
```

#### 9). isascii()

```python
s = input("Enter the string: ")
L = 0

for i in s:
    L+=1
if L == 0:
    print(False)
else:
    for i in s:
        ascii_val = ord(i)
        if not ((ascii_val >= 0 and ascii_val <= 127)):
            print(False)
            break
    else:
        print(True)
```

#### 10). startswith()

```python
s = input("Enter the string starts with:- ")
starts = input("Enter start letter you want to check:- ")

for i in s:
    if(s[0] in starts):
        print(True)
    else:
        print(False)
    break
```

#### 11). endswith()

```python
s = input("Enter the string ends with:- ")
ends = input("Enter start letter you want to check:- ")


for i in s:
    if(s[-1] in ends):
        print(True)
    else:
        print(False)
    break
```
