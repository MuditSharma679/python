### elif:-

#### Q.1 Take 3 input from user, and check whether which one is greater.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if(a == b == c):
    print("All three numbers are equal:", a)
elif(a >= b and a >= c):
    print("Greatest number is:", a)
elif(b >= a and b >= c):
    print("Greatest number is:", b)
else:
    print("Greatest number is:", c)
```

####  Take 3 input from user, and check whether which one is smaller.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if(a == b == c):
    print("All three numbers are equal:", a)
elif(a <= b and a <= c):
    print("Smallest number is:", a)
elif(b <= a and b <= c):
    print("Smallest number is:", b)
else:
    print("Smallest number is:", c)
    
```

#### Q.2 Enter marks of students for 5 subjects. Find out total marks based on this find out percentage and based on percentage show grades.
Grades- 90-100 = A+
        80-90  = A
        70-80  = B+
        60-70  = B
        50-60  = C+
        40-50  = C
        36-40  = D
        Less than 36 = Fail
    

```python
English = int(input("English Marks - "))
Hindi = int(input("Hindi Marks - "))
Maths = int(input("Maths Marks - "))
Science = int(input("Science Marks - "))
Soc_Science = int(input("Science Marks - "))
Total_Marks = 500

Obtained_Marks = English + Hindi + Maths + Science + Soc_Science
print("The Obtained marks is ",Obtained_Marks, "out of ",Total_Marks)

Percentage = (Obtained_Marks/Total_Marks)*100
print("The Percentage get ",round(Percentage,2),"%")

if(Percentage > 90 and Percentage <= 100 ):
    print("You got A+ Grade")
elif(Percentage > 80 and Percentage <= 90 ):
    print("You got A Grade")
elif(Percentage > 70 and Percentage <= 80 ):
    print("You got B+ Grade")    
elif(Percentage > 60 and Percentage <= 70 ):
    print("You got B Grade")
elif(Percentage > 50 and Percentage <= 60 ):
    print("You got C+ Grade")    
elif(Percentage > 40 and Percentage <= 50 ):
    print("You got C Grade")    
elif(Percentage > 36 and Percentage <= 40 ):
    print("You got D Grade")
else:
    print("---BAD LUCK FAIL---")
    
```
