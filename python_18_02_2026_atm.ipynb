### Nested Loop and If-elif-else

```python
num = int(input("Enter a number: "))

if(num>=0 and num%2 == 0):
    print("Number is positive and Even")
elif(num<0 and num%2 == 0):
    print("Number is negative and Even")
elif(num>=0 and num%2 != 0):
    print("Number is Positive and ODD")
elif(num<0 and num%2 != 0):
    print("Number is Negative and ODD")
else:
    print("Invalid Input")
    
```

```python
# Nested COndition for Even, Odd and Positive, Negative Combine.
num = int(input("Enter a number: "))

if(num == 0):
    print(num," is a nuetral number")
elif(num>0):
    if(num%2 == 0):
        print(num," is a Positive and Even Number")
    else:
        print(num," is a Positive and Odd Number")
else:
    if(num%2 == 0):
        print(num," is a Negative and Even Number")
    else:
        print(num," is a Negative and Odd Number")
    
```

Q.1 Take number of tyre as input for the following Conditions:

```python
tyres = int(input("Enter number of tyres: "))

if(tyres <= 0):
    print("Invalid Input")
elif(tyres < 5):
    if(tyres == 1):
        print("Unicycle")
    elif(tyres == 2):
        print("2-Wheeler/Bike")
    elif(tyres == 3):
        print("3-Wheeler/Auto")
    elif(tyres == 4):
        print("4-Wheeler/Car")

elif( tyres <= 50):
    if(tyres%2==0):
        print("Heavy Vehicles")
    else :
        print("Invalid Format")   
else:
    print("Invalid Input")
    
```

Q.2 ATM Pin check - correct/incorrect, balance check, balance withdraw , minimum_balance  = 1000.  

```python
pin = int(input("Enter Your PIN:- "))
balance = 10000

if(pin == 1234):
    print("CORRECT ATM PIN")
    print("Your Balance: ",balance)
    withdraw_amount = int(input("Enter the Amount to Withdraw:- "))
    if(balance >= withdraw_amount):
        if(balance <= (withdraw_amount+1000)):
            print(withdraw_amount,"RS failed to withdraw because it is "+str(1000-(balance-withdraw_amount))+" more than minimum balance")
        else :
            balance = balance - withdraw_amount
            print(withdraw_amount,"RS withdraw successfully")
            print("Remaining Balance is:-",balance)
    
else:
    print("Incorrect PIN")
```
