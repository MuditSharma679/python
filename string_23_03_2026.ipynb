# STRING :- 

String represents a group of characters.
String are enclosed in single or doule quotes.
The 'str' datatype represents a string.

```python
s = 'hello'
print(s)
type(s)
```

We can create string in 8 ways:-

```python
#(i) Single Quotes 
a = 'hello'
print("Single Quotes:- ",a)

#(ii) Double Quotes
b = "hello"
print("Double Quotes:- ",b)

#(iii) Triple Single Quotes/Double Quotes:-  We use it when we want to print statement in multiple lines or it prints as it is in the string.
c = '''hello 
                 world'''
print("Triple Single Quotes:- ",c)
g = """hello 
                 world"""
print("Triple Double Quotes",g)


#(iv) Double quotes inside Single quotes 
d = 'hello my name is "Mudit".'
print(d)

#(v) Single quotes inside Double quotes 
e = "hello my name is 'Mudit'."
print(e)

#(vi)Using Escape Character:- like /n = newline, /t = tab


#(vii)Raw String :-
print("Raw String:- ",r"hello my name is /t Mudit")



```

```python
# So conclusion is if two variable has same value, then they have same memory address.
s1 = "hello"
s2 = "hello"
print(id(s1))
print(id(s2))
```

Q.2) Take a string from user & find out for each word whether it is vowel or consonant.
    And also find out the count of Vowel and Consonant.

```python
s = input("Enter string:- ")
count_vow = 0
count_cons = 0

for i in s:
    if(i in ('a','e','i','o','u')):
        print(i," is a vowel")
        count_vow+=1
    else:
        print(i," is not a vowel")
        count_cons+=1
print("Count of Vowel is:- ",count_vow)
print("Count of Consonant is:- ",count_cons)
    
```

Q.3) Take a string from user & find out for each word whether which one is Capital or Small. And also find out the count of Capital and Small.

```python
# # ASCII value nikalne ke liye ord() use karein, chr() se hum ascii value ka digit findout krege.
s = input("Enter the string:- ")
count_capital=0
count_small=0
count_invalid=0

for i in s:
    
    ascii_val = ord(i)
    
    if(ascii_val >= 65 and ascii_val <= 90):
        print(i," is a CAPITAL LETTER")
        count_capital+=1
    elif(ascii_val >= 97 and ascii_val <= 122):
        print(i," is small letter")
        count_small+=1
    else:
        print("Invalid")
        count_invalid+=1
        
print("Count of CAPITAL LETTERS IS:- ",count_capital)
print("Count of small letter is:- ",count_small)
print("Invalid Count:- ",count_invalid)
        
```

```python
s = input("Enter the string:- ")

count_capital = 0
count_small = 0
count_vow = 0
count_cons = 0
count_invalid = 0

count_cap_vowel = 0
count_cap_cons = 0
count_small_vowel = 0
count_small_cons = 0

count_digits = 0
count_even = 0
count_odd = 0
sum_digits = 0

count_special = 0

for i in s:
    ascii_val = ord(i)

    #For CAPITAL LETTER check
    if(ascii_val >= 65 and ascii_val <= 90):
        count_capital += 1
        print(i, "is a CAPITAL LETTER", end=" ")

        if i in ('A', 'E', 'I', 'O', 'U'):
            print("and a Vowel")
            count_vow += 1
            count_cap_vowel += 1
        else:
            print("and a Consonant")
            count_cons += 1
            count_cap_cons += 1

    # For small letter check
    elif(ascii_val >= 97 and ascii_val <= 122):
        count_small += 1
        print(i, "is a small letter", end=" ")

        if i in ('a','e','i','o','u'):
            print("and a Vowel")
            count_vow += 1
            count_small_vowel += 1
        else:
            print("and a Consonant")
            count_cons += 1
            count_small_cons += 1

    # For digit check
    elif(ascii_val >= 48 and ascii_val <= 57):
        count_digits += 1
        digit = int(i)
        sum_digits += digit

        print(i, "is a Digit", end=" ")

        if(digit % 2) == 0:
            print("and EVEN")
            count_even += 1
        else:
            print("and ODD")
            count_odd += 1

    else:
        print(i, "is a Special Character")
        count_special += 1

print("\n---------------- Final Output ----------------")
print("Capital Letters:   ", count_capital,end=" | ")
print("Small Letters:     ", count_small)
print("-----------------------------------------------")
print("Vowels:            ", count_vow,end=" | ")
print("Consonants:        ", count_cons)
print("-----------------------------------------------")
print("Capital Vowels:    ", count_cap_vowel,end=" | ")
print("Capital Consonants:", count_cap_cons)
print("-----------------------------------------------")
print("Small Vowels:      ", count_small_vowel,end=" | ")
print("Small Consonants:  ", count_small_cons)
print("-----------------------------------------------")
print("Even Digits:       ", count_even,end=" | ")
print("Odd Digits:        ", count_odd)
print("Total Digits:      ", count_digits,end=" | ")
print("Sum of Digits:     ", sum_digits)
print("-----------------------------------------------")
print("No. of Special Characters:", count_special)
```

 Q.2) Take a string from user & find out for each word whether it is vowel or consonant. And also find out the count of Vowel and Consonant.

```python
s = input("Enter string:- ")

count_vow = 0
count_cons = 0
i = 0 

while(i < len(s)):   
    ch = s[i]
    if(ch in ('a','e','i','o','u')):
        print(ch, "is a vowel")
        count_vow += 1
    else:
        print(ch, "is not a vowel")
        count_cons += 1
    i += 1  

print("Count of Vowel is:- ", count_vow, end=" | ")
print("Count of Consonant is:- ", count_cons)
```
