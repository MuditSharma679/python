Q.1 WAP using match case to print day name by taking number as input from user.

```python
day_num = int(input("Enter the day number:- "))

match(day_num):
    case 1:
        print("Monday")
    case 2:
        print("Tuesday") 
    case 3:
        print("Wednesday")
    case 4:
        print("Thrusday")
    case 5:
        print("Friday")    
    case 6:
        print("Saturday") 
    case 7:
        print("Sunday")
    case __:
        print("Invalid")   
        
        
        
```

Q.3 WAP to check whether the letter is vowel or not.

```python
alphabet = input("Enter the letter from Alphabet:- ").lower()


match(alphabet):
    case "a" | "e" | "i" | "o" | "u" :
        print(" is a vowel")
    case __:
        print("Not a vowel")
```

Q.4 WAP where take 2 inputs from user and give choice for arithmetic operators by pressing 1,2,3,4.

```python
fir_num = int(input("Enter the first number:- "))
sec_num = int(input("Enter the second number:- "))

print("Press 1 for Add(+)")
print("Press 2 for Sub(-)")
print("Press 3 for Multi(*)")
print("Press 4 for Div(%)") 

choice = int(input("Enter your choice:- "))

match(choice):
    case 1:
        print("Result Addition:- ",fir_num,"+",sec_num," = ",fir_num + sec_num)
    case 2:
        print("Result Addition:- ",fir_num,"-",sec_num," = ",fir_num - sec_num)
    case 3:
        print("Result Addition:- ",fir_num,"*",sec_num," = ",fir_num * sec_num)
    case 4:
        print("Result Addition:- ",fir_num,"/",sec_num," = ",fir_num / sec_num)
    case _:
        print("Invalid Choice")
        
```

Q.4 WAP where take 2 inputs from user and give choice for arithmetic operators by pressing +,-,*,/.

```python
fir_num = int(input("Enter the first number:- "))
sec_num = int(input("Enter the second number:- "))

print("Press 1 for Add(+)")
print("Press 2 for Sub(-)")
print("Press 3 for Multi(*)")
print("Press 4 for Div(%)") 

operator = input("Enter the operator:- ")

match(operator):
    case "+":
        print("Result Addition:- ",fir_num,"+",sec_num," = ",fir_num + sec_num)
    case "-":
        print("Result Addition:- ",fir_num,"-",sec_num," = ",fir_num - sec_num)
    case "*":
        print("Result Addition:- ",fir_num,"*",sec_num," = ",fir_num * sec_num)
    case "/":
        print("Result Addition:- ",fir_num,"/",sec_num," = ",fir_num / sec_num)
    case _:
        print("Invalid Choice")
        
```
